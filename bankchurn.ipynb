```python
import pandas as pd
```

```python
df = pd.read_csv('BankChurners.csv', usecols=['Attrition_Flag','Total_Trans_Ct','Total_Revolving_Bal','Total_Trans_Amt', 
                                             'Total_Ct_Chng_Q4_Q1','Total_Amt_Chng_Q4_Q1','Customer_Age',
                                             'Total_Relationship_Count','Credit_Limit', 'Months_on_book'])
```

```python

df.head(5)
```

```python
from pycaret.classification import*
s = setup(data=df, target = 'Attrition_Flag', session_id = 123)
```

```python
rf = create_model('rf')
```

```python
plot_model(rf, plot = 'confusion_matrix')
```

```python
rf_finalized = finalize_model(rf)
```

```python
save_model(rf_finalized,'rf_saved_12082021')
```
